```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import pandas as pd
import urls
import random
```

```python
stocks = {'sava':0,
          'surf':1,
          'crdf':2,
          'avxl':3}
```

```python
def extract_topic():
    
    df = pd.DataFrame(columns=['Stock', 'Sep 2020', 'Oct 2020', 'Nov 2020', 'Dec 2020', 'Jan 2021', 'Feb 2021',
                           'Mar 2021', 'Apr 2021', 'May 2021', 'Jun 2021', 'Jul 2021', 'Aug 2021', 'Sep 2021'])
    
    for key in stocks.keys():
        print("Executing for stock: ", key)
        url_list = urls.url_dict[key]

        df = pd.concat([df, pd.DataFrame({}, index=[stocks.get(key)], columns=df.columns)])
        df.iloc[stocks.get(key), 0] = key

        service = Service('path/to/chromedriver')
        driver = webdriver.Chrome(service=service)

        for i, url in enumerate(url_list):
            if url!="NA":
                driver.get(url)
                driver.implicitly_wait(random.randint(15, 20))

                text = ""
                paragraphs = driver.find_elements(By.CLASS_NAME, 'press-release-header__title')
                for p in paragraphs:
                    text += p.text
            else:
                text = "Neutral"
            df.iloc[stocks.get(key), i+1] = text
            print("Extracted text for ", df.columns[i+1])

        driver.quit()
        print("Shape of data after this iteration: ", df.shape ,"\n")
        
    df.columns = [f'{col}_topic' if i >= 0 else col for i, col in enumerate(df.columns)]
        
    return df
```

```python
def extract_content():
    
    df = pd.DataFrame(columns=['Stock', 'Sep 2020', 'Oct 2020', 'Nov 2020', 'Dec 2020', 'Jan 2021', 'Feb 2021',
                           'Mar 2021', 'Apr 2021', 'May 2021', 'Jun 2021', 'Jul 2021', 'Aug 2021', 'Sep 2021'])
    
    for key in stocks.keys():
        print("Executing for stock: ", key)
        url_list = urls.url_dict[key]

        df = pd.concat([df, pd.DataFrame({}, index=[stocks.get(key)], columns=df.columns)])
        df.iloc[stocks.get(key), 0] = key

        service = Service('path/to/chromedriver')
        driver = webdriver.Chrome(service=service)

        for i, url in enumerate(url_list):
            if url!="NA":
                driver.get(url)
                driver.implicitly_wait(random.randint(20, 25))
                text = ""
                paragraphs = driver.find_elements(By.TAG_NAME, 'p')
                for p in paragraphs:
                    text += p.text
            else:
                text = "Neutral"
            df.iloc[stocks.get(key), i+1] = text
            print("Extracted text for ", df.columns[i+1])

        driver.quit()
        print("Shape of data after this iteration: ", df.shape ,"\n")
        
    df.columns = [f'{col}_content' if i > 0 else col for i, col in enumerate(df.columns)]
        
    return df
```

```python
df_topic = extract_topic()
```

```python
df_content = extract_content()
```

```python
df = df_topic.merge(df_content, on ='Stock')
df.head()
```

```python
df = df.set_index('Stock')
df = df.reindex(columns=sorted(df.columns))
df.head()
```
